#### Import Essential Libraries and load Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Walmart Data Analysis and Forcasting.csv")
data.head()
```

#### Now we perform Gathering info about data, use descriptive Statistic and checking null values

```python
data.info()
```

```python
data.describe()
```

#### What is the average weekly sales for each store?

```python
avg_week_sales = data.groupby('Store')['Weekly_Sales'].mean()
avg_week_sales
```

```python
top_10 =  avg_week_sales.nlargest(10)
plt.figure(figsize=(10, 6))
top_10.plot(kind = 'bar', color = 'skyblue')
plt.title("Top 10 Stores by Average Weekly Sales")
plt.xlabel("Store")
plt.ylabel("Average Weekly Sales")
plt.show()
```

#### 2. What is the total weekly sales for all stores?


```python
avg_week_sales = data['Weekly_Sales'].sum()
print(f"The total weekly sales for all stores are:{avg_week_sales}")
```

#### 3. What is the standard deviation of weekly sales for all stores?


```python
std_sales = data['Weekly_Sales'].std()
print(f"The standard deviation of the weekly sales for all stores is:{std_sales}")
```

#### 4) Violin Plot of Weekly Sales

```python
plt.figure(figsize =(10, 6))

sns.violinplot(x = data['Weekly_Sales'], color = 'skyblue')
plt.title('Violin Plot of Weekly Sales')
plt.xlabel('Weekly Sales')


plt.show()
```

"This box plot visualizes the distribution and standard deviation of weekly sales. The line in the center of the box represents the median of the data, while the lower and upper edges of the box represent the first quartile (Q1) and the third quartile (Q3) of the data, respectively. The outer lines show the spread of the data and outliers beyond the outer lines. As the standard deviation increases, the distribution of the data and the size of the box will also widen. Outliers may be observed as data points beyond the outer lines."






```python
plt.figure(figsize = (8, 6))
plt.hist(data['Weekly_Sales'], bins = 20 , color = 'royalblue', edgecolor = 'black')
plt.xlabel('Weekly Sales')
plt.ylabel('Frequency')
plt.title('Distribution of Weekly Sales')
plt.show()
```

This histogram graph shows the distribution of weekly sales and helps us understand the density of the data. The larger the standard deviation, the wider and more dispersed the data distribution will be.

#### 5) How many holiday weeks are there in the dataset?


```python
holiday = data[data['Holiday_Flag']==1]['Holiday_Flag'].count()
print(f"Numbers of Holiday weeks:",holiday)
```

```python
holiday = data[data['Holiday_Flag']==1]['Holiday_Flag'].count()
plt.figure(figsize =(10, 6))
plt.bar(['Holiday Weeks', 'Non-Holiday Weeks'], [holiday, len(data)- holiday], color = ['royalblue', 'lightgray'] )
plt.xlabel('Week Type')
plt.ylabel('Number of weeks')
plt.title('Number of holidays Week - Non-Numbers of holidays Week')
plt.show()
```

#### 6. What is the total weekly sales during holiday weeks?

 

```python
holiday_sales = data[data['Holiday_Flag'] == 1]['Weekly_Sales'].sum()
print("Total weekly sales during holiday weeks:", holiday_sales)
```

```python
non_holiday_Sales = data[data['Holiday_Flag'] == 0]['Weekly_Sales'].sum()
plt.figure(figsize=(10, 6))
labels = ['Holidays Weeks', 'Non-Holidays weeks']
sizes = [holiday_sales, non_holiday_Sales]
colors = ['royalblue', 'lightskyblue']
explode = [0.1, 0]
plt.pie(sizes, explode = explode, labels = labels, colors = colors, autopct = '%1.1f%%', shadow = True,  startangle=90)
plt.axis('equal')
plt.title('Percentage of Total Weekly Sales during Holiday Weeks vs. Non-Holiday Weeks')
plt.show()
```

#### 7. What is the correlation between weekly sales and temperature?

```python
cor = data['Weekly_Sales'].corr(data['Temperature'])
print("Correlation between weekly sales and temperature:", cor)

```

```python
plt.figure(figsize=(10, 8))
sns.scatterplot(data, x = 'Temperature', y = 'Weekly_Sales', color =  'skyblue', alpha = 0.7)

sns.regplot(data= data, x = 'Temperature', y = 'Weekly_Sales', color = 'red', scatter = False)

plt.xlabel('Temperature')
plt.ylabel('Weekly Sales')
plt.title('Weekly Sales vs. Temperature with Regression Line')
plt.grid(True)
plt.show()
```

 Each point represents weekly sales and temperatures. Additionally, the regression line illustrates a possible relationship between temperature and weekly sales. If the points are close to the line, there may be a positive correlation between temperature and sales. Conversely, if the points are far from the line, there might be a less pronounced relationship between the two variables.

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(data=data, x='Temperature', y='Weekly_Sales', hue='Weekly_Sales', palette='viridis', alpha=0.7)

plt.xlabel('Temperature')
plt.ylabel('Weekly Sales')
plt.title('Scatter Plot of Weekly Sales vs. Temperature')
plt.grid(True)
plt.show()
```

Each point represents an observation, where the x-axis represents the temperature value and the y-axis represents the value of weekly sales. Additionally, the color of the points represents the weekly sales, and the color scale varies according to the sales values. In this way, we can visualize the relationship between temperature and weekly sales in more detail.






#### 8 What is the correlation between weekly sales and fuel price?

```python
cor = data['Weekly_Sales'].corr(data['Fuel_Price'])
print("Correlation between weekly sales and fuel price:", cor)
```

The correlation coefficient between weekly sales and fuel price is approximately 0.009. Since the correlation coefficient is very close to zero, it indicates a very weak linear relationship between these two variables.






#### 9. Correlation between weekly sales and CPI

```python
corr_coeff = data['Weekly_Sales'].corr(data['CPI'])
print("Correlation between Weekly_Sales and CPI:", corr_coeff)
```

#### General correlation matrix

```python
corr = data[['Weekly_Sales', 'Holiday_Flag', 'Temperature', 'Fuel_Price', 'CPI', 'Unemployment']].corr()

cmap = sns.color_palette("BuGn", as_cmap =  True)

fig, ax = plt.subplots(figsize=(10, 8))
sns.heatmap(corr, annot = True, cmap= cmap, ax = ax)
ax.set_title('General correlation matrix')
plt.show()     
```

The variable Weekly_Sales has a positive correlation with the variable Holiday_Flag. This indicates that sales increase on holidays.

The variable Weekly_Sales has a negative correlation with the variables Temperature, Fuel_Price, CPI, and Unemployment. This indicates that sales decrease as temperature, fuel price, consumer price index, and unemployment rate increase.

The variable Weekly_Sales is not related to the variable General_correlation_matrix. This indicates that the overall correlation matrix does not affect sales.

There are various degrees of positive or negative correlations among other variables. For example, there is a positive correlation between Temperature and Fuel_Price; there is a negative correlation between CPI and Unemployment.
